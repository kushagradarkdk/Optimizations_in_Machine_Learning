```
######### Answer to Question 4 ##########
####### Begin #######

## Logistic Loss ##
n=100
m=10
def get_dataset(n,m):
  X = np.random.randn(n,m)
  y = np.random.randn(n,1)
  for i in range(0,len(y)):
    if y[i]>=0.5:
      y[i]=1
    else:
      y[i]=-1
  # print(len(X))
  # print(w.shape)
  return X,y
```

```
#Logistic Loss Function#
def LogisticLossFun_naive(w,X,y,lam):
  f=0
  [n,m]= np.shape(X)
  g=np.zeros(m)
  for i in range(n):
    value=0
    for j in range(m):
      value = value + X[i][j]*w[j]
    t = value*y[i]
    f += np.logaddexp(0,-t)
    g += -(y[i]*X[i])/(1+np.exp(t));
  f = f + lam * (np.linalg.norm(w))**2
  g = g + 2 *lam* np.linalg.norm(w)
  return [f,g]
```

```
def LogisticLossFun_Vector(w,X,y,lam):
  [r,q]= np.shape(X)
  w=w.reshape(q,1)
  value = np.dot(X,w)*y
  f = np.sum(np.logaddexp(0,-value))
  yx = y*X
  g=np.zeros(q)
  v = 1+np.exp(value)
  g = -np.sum(np.divide(yx,v),axis=0)
  f = f + lam * (np.linalg.norm(w))**2
  g = g + 2 *lam* np.linalg.norm(w)
  return [f,g]
```

```

def numericalGrad(funObj, w, epsilon):
	m=len(w)
	grad=np.zeros(m)
	for i in range(m):
		wp=np.copy(w)
		wn=np.copy(w)
		wp[i] = w[i] + epsilon
		wn[i] = w[i] - epsilon
		grad[i] += (funObj(wp) - funObj(wn))/(2*epsilon)
	return grad
```

```
m=10
n=100
X,y = get_dataset(n,m)
w = np.random.randn(m)
import time
starting_time_naive = time.time()
loss_naive,grad_naive = LogisticLossFun_naive(w,X,y,1)
print(loss_naive)
print(grad_naive)
print("Naive Time Taken : ", time.time()-starting_time_naive )

starting_time_vector = time.time()
loss_naive,grad_naive = LogisticLossFun_Vector(w,X,y,1)
print(loss_naive)
print(grad_naive)
print("Vector Time Taken : ", time.time()-starting_time_vector )

starting_time_numgrad = time.time()
funObj = lambda w: LogisticLossFun_Vector(w,X,y,1)[0]
print(funObj(w))
print(numericalGrad(funObj,w,0.0001))
print("Numerical Time Taken : ",time.time()-starting_time_numgrad)
```

```
#HingeLoss
#Solution 2

def HingeLoss_naive(w,X,y,lam):
  f=0
  [n,m]= np.shape(X)
  g=np.zeros(m)
  for i in range(n):
    value=0
    for j in range(m):
      value = value + X[i][j]*w[j]
    t = value*y[i]
    f +=  max(0,1-t)
    g += -int(1-t>0)*y[i]*X[i]
  f = f + lam * (np.linalg.norm(w))**2
  g = g + 2 *lam* np.linalg.norm(w)
  return [f,g]
```

```
def HingeLoss_Vector(w,X,y,lam):
  [r,q]= np.shape(X)
  w=w.reshape(q,1)
  value = np.dot(X,w)*y
  f = np.sum(np.where(1-value<0,0,1-value),axis=0)
  g = np.sum(np.where(1-value<0,0,-y*X), axis=0)
  f = f + lam * (np.linalg.norm(w))**2
  g = g + 2 *lam* np.linalg.norm(w)
  return [f,g]
```

```
m=10
n=100
X,y = get_dataset(n,m)
w = np.random.randn(m)
import time
starting_time_naive = time.time()
loss_naive,grad_naive = HingeLoss_naive(w,X,y,1)
print(loss_naive)
print(grad_naive)
print("Time Taken : ", time.time()-starting_time_naive )

starting_time_vector = time.time()
loss_naive,grad_naive = HingeLoss_Vector(w,X,y,1)
print(loss_naive)
print(grad_naive)
print("Time Taken : ", time.time()-starting_time_vector )

starting_time_numgrad = time.time()
funObj = lambda w: HingeLoss_Vector(w,X,y,1)[0]
print(funObj(w))
print(numericalGrad(funObj,w,0.0001))
print("Time Taken : ",time.time()-starting_time_numgrad)
```

```
#Simple 2 layer function
#Solution 3

def TwoLayer_naive(w,X,y,lam):
  f=0
  [n,m]= np.shape(X)
  g=np.zeros(m)
  for i in range(n):
    value=0
    for j in range(m):
      value = value + X[i][j]*w[j]
    #t = value*y[i]
    t= y[i] - max(0,value)
    f +=  (t)**2
    #g += -int(1-t>0)*y[i]*X[i]
    g += -2*(t)*int(value>0)*X[i]
  f = f + lam * (np.linalg.norm(w))**2
  g = g + 2 *lam* np.linalg.norm(w)
  return [f,g]
```

```
def TwoLayer_Vector(w,X,y,lam):
  [r,q]= np.shape(X)
  w=w.reshape(q,1)
  value = np.dot(X,w)
  f = np.sum(np.where(value<0,y**2,(y-value)**2),axis=0)
  g = np.sum(np.where(value<0,0,-2*(y-value)*X), axis=0)
  f = f + lam * (np.linalg.norm(w))**2
  g = g + 2 *lam* np.linalg.norm(w)
  return [f,g]
```

```
m=10
n=100
X,y = get_dataset(n,m)
w = np.random.randn(m)
import time
starting_time_naive = time.time()
loss_naive,grad_naive = TwoLayer_naive(w,X,y,1)
print(loss_naive)
print(grad_naive)
print("Time Taken : ", time.time()-starting_time_naive )

starting_time_vector = time.time()
loss_naive,grad_naive = TwoLayer_Vector(w,X,y,1)
print(loss_naive)
print(grad_naive)
print("Time Taken : ", time.time()-starting_time_vector )

starting_time_numgrad = time.time()
funObj = lambda w: TwoLayer_Vector(w,X,y,1)[0]
print(funObj(w))
print(numericalGrad(funObj,w,0.0001))
print("Time Taken : ",time.time()-starting_time_numgrad)
```

```
#Least Square Loss
# Solution 4
def LeastSquareLoss_naive(w,X,y,lam):
  f=0
  [n,m]= np.shape(X)
  g=np.zeros(m)
  for i in range(n):
    value=0
    for j in range(m):
      value = value + X[i][j]*w[j]
    #t = value*y[i]
    t= y[i] - value
    f +=  (t)**2
    #g += -int(1-t>0)*y[i]*X[i]
    g += -2*(t)*X[i]
  f = f + lam * (np.linalg.norm(w))**2
  g = g + 2 *lam* np.linalg.norm(w)
  return [f,g]
```

```
def LeastSquareLoss_Vector(w,X,y,lam):
  [r,q]= np.shape(X)
  w=w.reshape(q,1)
  value = np.dot(X,w)
  f = np.sum((y-value)**2,axis=0)
  g = np.sum(-2*(y-value)*X, axis=0)
  f = f + lam * (np.linalg.norm(w))**2
  g = g + 2 *lam* np.linalg.norm(w)
  return [f,g]
```

```
m=10
n=100
X = np.random.randn(n,m)
y = np.random.rand(n, 1)
w = np.random.randn(m)
import time
starting_time_naive = time.time()
loss_naive,grad_naive = LeastSquareLoss_naive(w,X,y,1)
print(loss_naive)
print(grad_naive)
print("Time Taken : ", time.time()-starting_time_naive )

starting_time_vector = time.time()
loss_naive,grad_naive = LeastSquareLoss_Vector(w,X,y,1)
print(loss_naive)
print(grad_naive)
print("Time Taken : ", time.time()-starting_time_vector)

starting_time_numgrad = time.time()
funObj = lambda w: LeastSquareLoss_Vector(w,X,y,1)[0]
print(funObj(w))
print(numericalGrad(funObj,w,0.0001))
print("Time Taken : ",time.time()-starting_time_numgrad)
```
